## Import data

```python
# URL
url = "https://www.basketball-reference.com/leagues/NBA_2022_per_game.html"
```

```python
# Read html webpage into pandas
import pandas as pd
df = pd.read_html(url, header = 0)  #first row is the header
print(df)
```

```python
# Know more about our webpage

# How many tables
print(f"Total tables: {len(df)}")
# Create table 2022
df2022 = df[0]
# Colums in the dataframe
print(df2022.columns)
```

## Data Cleaning 
We can see the sub header should be removed
Screen Shot 2022-03-04 at 10.40.20 PM.png

```python
df2022['Pos'].unique()
```

```python
print(f"There are total of {len(df2022[df2022.Pos == 'Pos'])} sub header in our table.")
# Visualize the the duplicate sub header
df2022[df2022.Pos == 'Pos']
```

```python
# Remove the duplicate sub header (794-29=765)
df2022 = df2022.drop(df2022[df2022.Pos == 'Pos'].index)
print(df2022.shape)
```

## Some question to do EDA
- Which player scored the most points?
- Explore some information of that player (position, team, points, number of games, etc)
- Which player scored the most points for 3 point field goals? 
- Total points of each team
- Do some EDA on one of the teams (PHI) 



```python
df2022.columns
```

```python
df2022.describe()
```

```python
#Display the row that has player has the highest point
df2022[df2022['PTS'] == df2022.PTS.max()]
```

```python
# Information from that player
df2022[df2022['PTS'] == df2022.PTS.max()].Tm
```

```python
# Create a variable for that player instead of typing the whole command out like above
MaxPlayer = df2022[df2022['PTS'] == df2022.PTS.max()]
print(MaxPlayer.Player) #Name
print(MaxPlayer.PTS)    #Point
print(MaxPlayer.Tm)     #Team   
print(MaxPlayer.G)      #Total game
```

```python
# Player has the highest 3P
Highest3P = df2022[df2022['3P'] == df2022['3P'].max()]
Highest3P
```

```python
df2022['3P'].max()
```

```python
# Show players have lowest points and highest points (practice using logical operation)
df2022[(df2022['PTS'] == df2022['PTS'].min()) | (df2022['PTS'] == df2022['PTS'].max())]
```

```python
# Using groupby to see total point of each team
df2022.groupby('Tm')['PTS'].sum()
```

```python
# Plot total points of each team
df2022.groupby('Tm')['PTS'].sum().plot(kind='bar')
```

```python
# Create variable to store team PHI and do some EDA about that team
PHI = df2022.groupby('Tm').get_group('PHI')
PHI
```

```python
# Player has the highest score 
PHI[PHI.PTS == PHI.PTS.max()]
```

```python
# Summary statistics for PHI to see the contribution of each player
PHI.groupby('Player').PTS.describe()
```

```python
# Assist
PHI.groupby('Player').AST.describe()
```

```python
# Positions contribution
PHI.groupby('Pos').PTS.describe()
```
